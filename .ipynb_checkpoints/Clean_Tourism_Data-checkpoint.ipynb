```python
import pandas as pd
```

```python
# read dataset
df = pd.read_csv('db/tourism.csv')
df.head()
```

```python
# select tourism data: "Indicator Id" between 1837-1846
df = df[ (df['Indicator Id'] >= 1837) & (df['Indicator Id'] <= 1846) ]
df.head()
```

```python
# delete unwanted columns
del df['Indicator Id']
del df['Subindicator Type']

# unpivot the year columns
df2 = pd.melt(df, id_vars=['Country ISO3', 'Country Name', 'Indicator'],
              var_name = 'Year', value_name ='Value')
df2.head()
```

```python
# select years >= 2000
df2['Year'] = df2['Year'].astype(int)
df3 = df2[ df2['Year'] >= 2000 ]
df3.head()
```

```python
# pivot the Indicator
df4 = df3.pivot(columns='Indicator', values='Value')
df4.head()
```

```python
# join country names with pivot table based on index
df5 = df3.join(df4)

# delete unwanted columns
del df5['Value']
del df5['Indicator']
df5.head()
```

```python
# save dataset
df5.to_csv('db/tourism_cleaned.csv', index=False)
```
