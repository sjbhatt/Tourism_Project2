```python
import pandas as pd
```

```python
# read dataset
df = pd.read_csv('db/tourism.csv')
df.head()
```

```python
# select tourism data: "Indicator Id" between 1837-1846
df = df[ (df['Indicator Id'] >= 1837) & (df['Indicator Id'] <= 1846) ]
df.head()
```

```python
# delete unwanted columns
del df['Indicator Id']
del df['Subindicator Type']

# unpivot the year columns
df2 = pd.melt(df, id_vars=['Country ISO3', 'Country Name', 'Indicator'],
              var_name = 'Year', value_name ='Value')
df2.head()
```

```python
# select years >= 2000
df2['Year'] = df2['Year'].astype(int)
df3 = df2[ df2['Year'] >= 2000 ]
df3.head()
```

```python
# pivot the Indicator
df4 = df3.pivot_table(index=['Country ISO3', 'Country Name', 'Year'], columns='Indicator', values='Value').reset_index()
df4.columns.name = ''
df4.head()
```

```python
# rename fields into SQL format
df4_abbr = df4.rename(columns = {'Country ISO3': 'country_iso3',
                                 'Country Name': 'country_name',
                                 'Year': 'year',
                                 'International tourism, expenditures (% of total imports)': 'expd_pct_imp',
                                 'International tourism, expenditures (current US$)': 'expd_total',
                                 'International tourism, expenditures for passenger transport items (current US$)': 'expd_trans',
                                 'International tourism, expenditures for travel items (current US$)': 'expd_items',
                                 'International tourism, number of arrivals': 'arrivals',
                                 'International tourism, number of departures': 'departures',
                                 'International tourism, receipts (% of total exports)': 'rcpt_pct_exp',
                                 'International tourism, receipts (current US$)': 'rcpt_total',
                                 'International tourism, receipts for passenger transport items (current US$)': 'rcpt_trans',
                                 'International tourism, receipts for travel items (current US$)': 'rcpt_items'
                                 })
```

```python
# save dataset
df4_abbr.to_csv('db/intl_tourism.csv', index=False)
```
